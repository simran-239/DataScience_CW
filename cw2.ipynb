```python
#Q1 Finding the sum of numbers in a list using while loop
l=[12,35,67,78,23]
sums=0
i=0
while i < len(l):
    sums+=l[i]
    i+=1
print(sums)
```

```python
#Q2 Printing the square of numbers using while loop
i=0
while i<=10:
    print(i*i)
    i+=1
```

```python
#Q3 Finding the sum of even numbers using while loop
i=0
sums=0
while i<=10:
    if i%2==0:
        sums+=i
    i+=1
print(sums)
```
