```python
def power(a, b):
    if b == 0:
        return 1
    return a * power(a, b - 1)

print(power(2, 3))
```

```python
keys = ['Class-V', 'Class-VI', 'Class-VII', 'Class-VIII']
values = [1, 2, 2, 3]
class_dict = {}

# Populate the dictionary
for key, value in zip(keys, values):
    if key not in class_dict:
        class_dict[key] = set()
    class_dict[key].add(value)

# Print the dictionary
print(class_dict)
```

```python
#LINEAR ALGEBRA
matrix3= np.arange(1,10).reshape(3,3)
matrix4= np.eye(3,3)  #for identity matrix
print(matrix3)
print('\n')
print(matrix4)
```

```python
#Solve these equations for X,Y,Z
# X+ 2Y+ 3Z= 10
# 2X+ 3Y+ Z= 15
# X+ Y+ 2Z= 20
```

```python
# Coefficients of the equations
A = np.array([[1, 2, 3],[2, 3, 1],[1, 1, 2]])

# Constants on the right-hand side
B = np.array([10, 15, 20])

# Solve the system of equations
X,Y,Z = np.linalg.solve(A, B)
print("X =", X)
print("Y =", Y)
print("Z =", Z)
```
